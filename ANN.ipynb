```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, SimpleRNN, Embedding, LSTM, Bidirectional, InputLayer
from tensorflow.keras.optimizers import Adam
from keras.preprocessing import sequence
```

```python
waiter_df = pd.read_csv("Waiter_reward.csv")
level_df = pd.read_csv("Level.csv")
```

```python
df = pd.concat([waiter_df,level_df], axis=1)
```

```python
df.drop(['customer_satisfaction_score'], inplace=True, axis=1)
```

```python
df.head()
```

```python
X = df.drop(['Level', 'Waiter_id'], axis =1).values
```

```python
X.shape
```

```python
pd.get_dummies(df['Level'])
```

```python
y=pd.get_dummies(df['Level']).values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

# ANN

```python
ann_model = Sequential()

ann_model.add(Dense(4, activation='relu'))

# ann_model.add(Dropout(0.5))
ann_model.add(Dense(4, activation='relu'))

# ann_model.add(Dropout(0.5))
# ann_model.add(Dense(4, activation='relu'))

# ann_model.add(Dropout(0.5))
# ann_model.add(Dense(4, activation='relu'))

ann_model.add(Dropout(0.2))
ann_model.add(Dense(4, activation='softmax'))

ann_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
ann_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=400)
```

```python
ann_losses = pd.DataFrame(ann_model.history.history)
ann_losses
```

```python
plt.figure(figsize=(12,7))
plt.plot(ann_losses)
plt.legend(['loss', 'accuracy', 'val_loss', 'val_accuracy'])
```

# RNN 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
rnn_model = Sequential()
rnn_model.add(LSTM(units=4, return_sequences=True, input_shape=(X_test.shape[1], 1)))
rnn_model.add(Dropout(0.2))

rnn_model.add(LSTM(units=4, return_sequences=True ))
rnn_model.add(Dropout(0.2))
          
rnn_model.add(LSTM(units=4))
rnn_model.add(Dropout(0.2))
          
rnn_model.add(Dense(4, activation='softmax'))
        
rnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=rnn_model.fit(X_train, y_train, epochs=100, batch_size=35, validation_data=(X_test, y_test))
```

```python
rnn_losses = pd.DataFrame(rnn_model.history.history)
rnn_losses
```

```python
plt.figure(figsize=(12,7))
plt.plot(rnn_losses)
plt.legend(['loss', 'accuracy', 'val_loss', 'val_accuracy'])
```
